# Imports

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('blurry1.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('blurry2.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.imread('blurry3.jpg')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.imread('blurry4.jpg')
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
```

## Variance of Laplacian for blur detection
Analysis of focus measure operators for shape-from-focus 2013 Petruz et al.

```python
img1_ = cv2.Laplacian(img1, cv2.CV_64F).var()
```

set the variance of Laplacian threshold to 100 \
if the variance falls below 100 then it is blurry

```python
img1_
```

```python
img1_labeled = cv2.putText(img1, "{}: {:.2f}".format('Blurry:', img1_), (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 3)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img1_labeled)
```

```python
img3_ = cv2.Laplacian(img3, cv2.CV_64F).var()
```

```python
img3_
```

```python
img3_labeled = cv2.putText(img3, "{}: {:.2f}".format('Not Blurry', img3_),\
                           (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 3)
```
