```python
import cv2
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
img =cv2.imread('blurry3.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(img)
```

```python
mean, eigenvectors = cv2.PCACompute(img, np.mean(img, axis=0).reshape(1,-1))
```
